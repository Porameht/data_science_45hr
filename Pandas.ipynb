```python
# change number data within series
faculties = pd.Series(['humanities','science','engineering','pharmacy','nursing'])
print(faculties)

faculties[5] = 'dentisty'
p = [5,4,3,2,5]

faculties = faculties[p]
print(faculties)
```

```python
# del
faculties = pd.Series(['humanities','science','engineering','pharmacy','nursing'])
print(faculties)

del faculties[1]
print(faculties)
```

```python
# sequen of langauge
faculties = pd.Series(['humanities','science','engineering','pharmacy','nursing'])
print(faculties)

faculties = faculties.sort_values()
print(faculties)
```

```python
# reverse sequen of langauge
faculties = pd.Series(['humanities','science','engineering','pharmacy','nursing'])
print(faculties)

faculties = faculties.sort_values(ascending=False)
print(faculties)
```
