```python
import math
class point:
    def __init__(self):
        self.xcoor = 0
        self.ycoor = 0
    def mov(self, x, y):
        self.xcoor = x
        self.ycoor = y
    def dist(self, pnt):
        distance = math.sqrt((self.xcoor - pnt.xcoor)**2 + (self.ycoor - pnt.ycoor)**2)
        return distance
        
pnt1 = point()
pnt2 = point()
print(pnt1.dist(pnt2))

pnt2.mov(2, 4)
print(pnt1.dist(pnt2))
```
